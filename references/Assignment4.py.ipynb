```python
for column in df.columns:
    if column=="Age":
        df[column].fillna(24,inplace=True)
    elif column=="Cabin":
        df[column].fillna("C23 C25 C27", inplace=True)
    elif column=="Embarked":
        df[column].fillna("S", inplace=True)
```
