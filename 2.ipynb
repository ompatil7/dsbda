```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

miss = [0]
df = pd.read_csv('academicperformance.csv', na_values=miss)
df.head()
```

```python
df.columns
```

```python
cols = ['Femarks', 'SEMarks', 'Temarks', 'Bemarks']
df.boxplot(cols)
```

```python
plt.scatter(df['Femarks'], df['SEMarks'])
```

```python
df.dropna(inplace =True)
plt.scatter(df['Femarks'], df['SEMarks'])
```

```python
df
```

```python

q1_fe = df['Femarks'].quantile(0.25)
q3_fe = df['Femarks'].quantile(0.75)
iqr_fe = q3_fe - q1_fe
low_fe = q1_fe - 1.5 * iqr_fe
high_fe = q3_fe + 1.5 * iqr_fe
outlliers_fe = df[(df['Femarks'] < low_fe) | (df['Femarks'] > high_fe)]
outlliers_fe

df = df.drop(outlliers_fe.index)
```

```python
plt.scatter(df['Femarks'], df['SEMarks'])
```

```python
# data normalization

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[cols] = scaler.fit_transform(df[cols])
df
```

```python
dfnew = pd.read_csv('academicperformance.csv', na_values = [0])
dfnew.dropna(inplace = True)

def dec(col):
    max = np.max(np.abs(col))
    numdigit = len(str(int(max)))
    scale = 10**numdigit
    return col / scale

dfnew[cols] = dfnew[cols].apply(dec)
dfnew
```
