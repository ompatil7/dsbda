```python
#Load the Dataset into pandas data frame.
import pandas as pd

df = pd.read_csv('IRIS.csv')
df

df.head()
```

```python
print("Missing values in the dataset")
print(df.isnull())
print()
print("Summary of missing values")
print(df.isnull().sum())
print("Information about the dataset")
print(df.info())
```

```python
# Apply proper data type conversion
#Float to int
df['sepal_length'] = df['sepal_length'].astype(int)
#object to categorial
df['species'] = df['species'].astype('category')
print(df.head())
print(df.dtypes)
```

```python
#7 
from sklearn.preprocessing import LabelEncoder

# Use LabelEncoder to convert 'species' to numeric values
label_encoder = LabelEncoder()
df['species_encoded'] = label_encoder.fit_transform(df['species'])
#converts the values to 0, 1, 2,3..n categories
print(df['species_encoded'])
#label encoder does not create seperate col but can add bias in our model as 
#it will start giving higher preference to the Female parameter as 1>0 

# df.drop(df['species'], axis=1)
df
```

```python
#one hot encoding creates a seperate col for every category

#to print unique values of the cateogiral colummn
print(df['species'].unique())
print(df['species'].value_counts())#n values of each category

#onehotencoding
df['species'] = df['species'].astype('category')
df1 = pd.get_dummies(df['species'], dtype = int)
df1
#  it will create a new DataFrame where each unique value in the 'species' column becomes a new column,
# and for each row, these new columns will have a 1 if the original 'species' column had that value, and 0 otherwise.
dfnew = pd.concat([df, df1], axis=1)
dfnew
dfnew['species'].replace({'Iris-setosa': 0 ,'Iris-versicolor' : 1, 'Iris-virginica': 2}, inplace=True)
dfnew


```
